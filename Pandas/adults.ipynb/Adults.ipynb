```python
#the Most Common Race Among "Never-married"
def most_common_never_married_race(data):
    never_married = data[data['marital-status'] == 'Never-married']
    common_race = never_married['race'].mode()[0]
    return common_race
```

```python
#Percentage of People with a College Degree or Higher
def percentage_college_degree_or_higher(data):
    total_people = len(data)
    college_degree_or_higher = data[data['education'].isin(['Bachelors', 'Masters', 'Doctorate'])]
    percentage = (len(college_degree_or_higher) / total_people) * 100
    return percentage
```

```python
#New Column Named "age_group" that Categorizes People into Young (<=30), Middle-aged (31-60), and Senior (>60)
def categorize_age(age):
    if age <= 30:
        return 'Young'
    elif 31 <= age <= 60:
        return 'Middle-aged'
    else:
        return 'Senior'
   
```

```python
df1["age_group"] = df1['age'].apply(categorize_age)
print(df1)
```
